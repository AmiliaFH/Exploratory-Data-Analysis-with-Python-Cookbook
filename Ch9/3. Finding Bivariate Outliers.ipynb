#### Import relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Load dataset and subset columns

```python
houseprice_data = pd.read_csv("data/HousingPricesData.csv")
```

```python
houseprice_data = houseprice_data[['Price', 'Area', 'Room']]
```

#### Inspect first 5 rows and data types of the dataset

```python
houseprice_data.head()
```

```python
houseprice_data.shape
```

#### Generate Summary Statistics

```python
houseprice_data.describe().apply(lambda s: s.apply('{0:.1f}'.format))
```

#### Identify bivariate outliers using Scatter plot 

```python
plt.figure(figsize = (40,18))

ax = sns.scatterplot(data= houseprice_data, x= 'Price', y = 'Area',s = 200)
ax.set_xlabel('Price in millions', fontsize = 30)
ax.set_ylabel('Area',  fontsize = 30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)

ax.set_title('Bivariate Outlier Analysis of House prices and Area', fontsize = 40)
```

#### Identify bivariate outliers using Boxplot

```python
plt.figure(figsize = (40,18))

ax = sns.boxplot(data= houseprice_data,x = 'Room', y= 'Price')
ax.set_xlabel('Room', fontsize = 30)
ax.set_ylabel('Price in millions',  fontsize = 30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
ax.set_title('Bivariate Outlier Analysis of House prices and Rooms', fontsize = 40)
```
